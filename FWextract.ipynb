```python
import numpy as np
import os
import pandas as pd
import nibabel as nib
from tqdm import tqdm
```

```python
main_path = r'F:\Project_AD'
df_dir = pd.read_csv(os.path.join(main_path, 'Code', 'DTI', 'Subject_DKI.txt'), header=None, names=['Subject', 'Time'])
```

```python
df_subj = pd.DataFrame()

mask_path = os.path.join(main_path, 'Data_template\\MNI152')
mask = nib.load(mask_path + "\\MNI152_T1_1mm_Brain_FAST_seg.nii.gz")
mask_data = mask.get_fdata()

for index, row in tqdm(df_dir.iterrows()):
    subj = row['Subject']
    time_s = row['Time']
    dtifit_path = os.path.join(main_path, 'Data_prep', subj, 'DTI', time_s, 'FWDKIfit')
    if os.path.exists(dtifit_path):
        #dkifit_path = os.path.join(main_path, 'Data_prep', subj, 'DTI', time_s, 'DKIfit')
    
        img = nib.load(dtifit_path + "\\DTI_frac_ants.nii.gz")
        img_data = img.get_fdata()
        roi_values = img_data[mask_data == 3]
        subject_data = pd.DataFrame([np.mean(roi_values)])
        subject_data.index=pd.MultiIndex.from_product([[subj],[time_s]])
    
        df_subj = pd.concat([df_subj, subject_data], ignore_index=False)
        #print(dtifit_path)
```

```python
df_subj.index.names=['PTID','Time']
df_subj.to_excel("F:/Project_AD/Data_excel/ALPS/freewater_DTI_orig.xlsx", index=True, engine="openpyxl")

```
