```python
import numpy as np
import os
import pandas as pd
import nibabel as nib
from tqdm import tqdm
import shutil
```

```python
main_path = r'F:\Project_AD'
df_dir = pd.read_csv(os.path.join(main_path, 'Code', 'DTI', 'Subject_DKI.txt'), header=None, names=['Subject', 'Time'])
```

```python
def roivalue(img_data, mask_data):
    num_volumes = img_data.shape[3]
    subject_data = {}
    for i in range(num_volumes):
        volume = img_data[..., i]
        roi_values = volume[mask_data > 0]
        subject_data[i] = np.mean(roi_values)
    return subject_data
```

```python
df_subj = pd.DataFrame()

img_path = os.path.join(main_path, 'Data_prep')
mask_path = r'F:\Project_AD\Data_mask\Mask_alps'

for index, row in tqdm(df_dir.iterrows()):
    subj = row['Subject']
    time_s = row['Time']
    dtifit_path = os.path.join(img_path, subj, 'DTI', time_s, 'DTIfit_ANTs')
    if os.path.exists(dtifit_path):
        #dkifit_path = os.path.join(main_path, 'Data_prep', subj, 'DTI', time_s, 'DKIfit')
    
        img = nib.load(dtifit_path + "\\DTI_tensorants.nii.gz")
        img_data = img.get_fdata()

        maskh = os.path.join(mask_path, subj, time_s)
        mask_lp = nib.load(mask + "\\dti_L_proj_mask.nii.gz")
        mask_la = nib.load(mask + "\\dti_asso_mask.nii.gz")
        mask_rp = nib.load(mask + "\\dti_proj_mask.nii.gz")
        mask_ra = nib.load(mask + "\\dti_asso_mask.nii.gz")
        mask_lp_data = mask_lp.get_fdata()
        mask_la_data = mask_la.get_fdata()
        mask_rp_data = mask_rp.get_fdata()
        mask_ra_data = mask_ra.get_fdata()
    
        df_lp = roivalue(img_data, mask_lp_data)
        df_la = roivalue(img_data, mask_la_data)
        df_rp = roivalue(img_data, mask_rp_data)
        df_ra = roivalue(img_data, mask_ra_data)

        subject_data = pd.concat([pd.DataFrame([df_lp]), pd.DataFrame([df_la]), pd.DataFrame([df_rp]), pd.DataFrame([df_ra])], axis=1)
        subject_data.index=pd.MultiIndex.from_product([[subj],[time_s]])
    
        df_subj = pd.concat([df_subj, subject_data], ignore_index=False)
        #print(dtifit_path)

df_subj.index.names=['PTID','Time']
df_subj.to_excel("F:/Project_AD/Data_excel/ALPS/ALPS_DTI_orig.xlsx", index=True, engine="openpyxl")
```

```python
df_extract = pd.DataFrame()
df_extract['alps_l'] = df_subj.iloc[:, [0, 6]].mean(axis=1) / df_subj.iloc[:, [3, 11]].mean(axis=1)
df_extract['alps_r'] = df_subj.iloc[:, [12, 18]].mean(axis=1) / df_subj.iloc[:, [15, 23]].mean(axis=1)
df_extract['alps_mean'] = df_extract.iloc[:, [0, 1]].mean(axis=1)
df_extract.index.names=['PTID','Time']

df_extract.to_excel("F:/Project_AD/Data_excel/ALPS/ALPS_DTI_extract.xlsx", index=True, engine="openpyxl")
print('End')
```
